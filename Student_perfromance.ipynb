```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
data = pd.read_csv("students_data.csv")


X = data[['study_hours_per_week',
          'attendance_percent',
          'assignments_completed',
          'sleep_hours',
          'internet_usage_hours_per_day']]

y = data['final_score']

# Model
model = LinearRegression()
model.fit(X, y)

# Predictions
pred_score = model.predict(X)

print("Predicted score sample:", round(pred_score[0], 2))

# ---- Metrics ----
mae = mean_absolute_error(y, pred_score)
mse = mean_squared_error(y, pred_score)
rmse = np.sqrt(mse)
r2 = r2_score(y, pred_score)

print("Mean Absolute Error:", round(mae, 2))
print("Mean Squared Error:", round(mse, 2))
print("Root Mean Squared Error::", round(rmse, 2))
print("R2 Score:", round(r2, 2))

# ---- Histogram ----
plt.figure(figsize=(10, 6))
plt.hist(data["final_score"], bins=30, edgecolor='black',label="Student_Data")
plt.title("Distribution of Final Exam Score")
plt.xlabel("Final Exam Score")
plt.ylabel("Number of Students")
plt.grid(True,linestyle='--')
plt.show()

# ---- Scatter Plot using one feature (study hours) ----
plt.figure(figsize=(10, 6))
plt.scatter(data["study_hours_per_week"], y, edgecolor='black', label="Actual Score")

# regression line
single_feature_pred = np.poly1d(np.polyfit(data["study_hours_per_week"], y, 1))
plt.plot(data["study_hours_per_week"], single_feature_pred(data["study_hours_per_week"]),
         label="Trend Line",color="red")

plt.title("Relation Between Study Hours & Final Score")
plt.xlabel("Study Hours Per Week")
plt.ylabel("Final Score")
plt.legend()
plt.grid(True,color='gray',linestyle='--')
plt.show()


```
